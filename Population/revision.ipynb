```python
# line plot data['date']/ value
import datetime
# 1- create a figure using plt.subplots, fix figsize=(8,6)
# 2-plot
plt.figure(figsize=(8, 6))
fig = plt.plot(data["date"], data["value"])
```

```python
data = pd.read_csv(
  "./multivariate.csv", 
    sep=';'
)
```

```python
data
```

```python
# plot the distribution of Birth rate and Mortality rate in the same plot 
#data[data["Children per woman"] == 7.43]
#data[data["Children per woman"] == data["Children per woman"].min()]
plt.plot(data["Birth rate"])
plt.plot(data["Mortality rate"])
```

```python
plt.figure(figsize=(10,7), dpi=80)
sns.distplot(data["Birth rate"], kde=True, hist=False)
sns.distplot(data["Mortality rate"], kde=True, hist=False)
```

```python
# print the correlation matrix 
mat = data.corr()
mat
```

```python
# plot the heatmap using the correlationn matrix
sns.heatmap(mat)
```

```python
# using pivot_table , average,min,max( morality_rate, birth_rate ) for each continent  , comments 
piv = data.pivot_table(index='Continent',values=['Mortality rate', 'Birth rate'],aggfunc=['mean', 'min', 'max'],margins=True)
piv
```

```python
# Average pop and growth_rate  for each group in one code sentence. Use the groupby() method. 
data.groupby('Continent')['Growth rate','Pop'].mean()
```

```python
# Sort the DataFrame in descending order of the growth_rate
data.sort_values("Growth rate", ascending=False)
```

```python
# get the countries where less than 5
data[data["Mortality rate"] < 5].sort_values("Mortality rate")
```
